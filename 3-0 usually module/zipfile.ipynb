# zipfile
    - 文件压缩操作

```python
import zipfile
```

```python
# ZipFile 创建一个ZipFile对象，表示zip文件
# 参数
# file 文件路径或类文件对象
# model 打开模式。默认'r'(只读)。'w':写，没有则创建，有则覆盖。'a'：追加，
# compression 压缩方法，
# allowZip64 若文件大小超过2G，则设为True

# help(zipfile.ZipFile)

zf = zipfile.ZipFile('D:\\jupyter_notebook\\variable\\3-0 usually module\\test_zip.zip')
print(zf)
```

```python
# getinfo 获取ZipFile对象中指定文件的详细信息
print(zf.getinfo("test.txt"))
```

```python
# namelist 获取ZipFile文件中的文件列表
nl = zf.namelist()
print(nl)
```

```python
# extractall 解压当前ZipFile文档到指定路径
# 参数
# path 解压后的保存路径 （推荐使用绝对路径）

import os
res = zf.extractall(os.path.join(os.getcwd(), 'extract_zip'))
print(zf)
```

```python
# close 释放ZipFile对象
zf.close()
```

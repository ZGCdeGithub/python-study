# timeit
  - 计时器库

```python
import timeit
```

```python
#利用timeit调用代码，执行10000次，查看运行时间
t1 = timeit.timeit('[i for i in range(1000)]',number=100000)

print(t1)
```

```python
def fun():
    print('*'*30)
    for i in range(50):
        print(i, end=' ')
    print('')
    
#查看函数的运行时间
t1 = timeit.timeit(stmt=fun, number=10)
print('')
print(t1)
```

```python
s = '''
def fun(num):
    print('*'*20)
    for i in range(num):
        print(i, end='  ')
    print('')
'''

#step 负责建立执行环节
t2 = timeit.timeit('fun(num)', setup=s+'num=10', number=10)
print(t2)
```
